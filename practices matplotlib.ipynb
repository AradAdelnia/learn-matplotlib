```python
import matplotlib.pyplot as plt
import numpy as np
labels = ["Minecraft", "Call of Duty", "GTA V","Red Dead 2","FIFA 2018", "Witcher 3", "Super Mario"]
data1 = [10, 31, 32, 17, 10, 6, 14]
data2 = [300, 100, 185, 55, 24, 50, 58]
x = np.arange(len(labels))
width = 0.35
fig, ax = plt.subplots()
ax.bar(x - width/2, data1, width, label = "2010", color = "red")
ax.bar(x + width/2, data2, width, label = "2020", color = "black")

ax.set_ylabel("mizan frosh be milion")
ax.set_title("mizan frosh bazeha be faseley 10 sal")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt # matplotlib     
import numpy as np #numpy   


vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"] #(y)  
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."] #(x)   

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]]) #       


fig, ax = plt.subplots()
im = ax.imshow(harvest)

#                    :
ax.set_xticks(np.arange(len(farmers)), labels=farmers)                            
ax.set_yticks(np.arange(len(vegetables)), labels=vegetables)                    

#             :
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

#          :
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="black")

ax.set_title("Harvest of local farmers (in tons/year)") #  
fig.tight_layout()
plt.show() # 
```
