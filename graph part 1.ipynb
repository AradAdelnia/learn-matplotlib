```python
import matplotlib.pyplot as plt
import numpy as np
labels = ["a", "b", "c", "d", "e"]
data1 = [20,34,30,35,27]
data2 = [25,32, 34, 20, 25]
data1_std = [2, 3, 4, 1,2]
data2_std = [3, 5, 2, 3, 3]
width = 0.35
fig, ax = plt.subplots()
ax.bar(labels, data1, width, yerr=data1_std, label = "data1", color = "red")
ax.bar(labels, data2, width, yerr=data2_std,bottom = data1, label = "data2", color = "green")
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
labels = ["a", "b", "c", "d", "e"]
data1 = [20,34,30,35,27]
data2 = [25,32, 34, 20, 25]
x = np.arange(len(labels))
width = 0.35
fig, ax = plt.subplots()
ax.bar(x - width/2 ,data1, width, label = "data1", color = "red")
ax.bar(x + width/2 ,data2, width, label = "data2", color = "gray")

ax.set_ylabel("y")
ax.set_title("bar")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

# Compute areas and colors
N = 150
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r**2
colors = theta

fig = plt.figure()
ax = fig.add_subplot(projection='polar')


c = ax.scatter(theta, r, c=colors, s=area, cmap='hot', alpha=0.75)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='polar')
c = ax.scatter(theta, r, c=colors, s=area, cmap='magma', alpha=0.75)

ax.set_rorigin(-1.5)
ax.set_theta_zero_location('W', offset=10)
plt.show()
```
